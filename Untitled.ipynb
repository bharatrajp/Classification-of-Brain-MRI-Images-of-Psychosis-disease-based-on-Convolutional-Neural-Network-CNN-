```python
df['Gender']=df['Gender'].map({'F':0,'M':1})
#here i change the gender (categorical feature to numericla ), where Male is 1 and female is 0. Using map function.
```

```python
df.head()
```

```python
# handling the categorical feature age 
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
# here i have done ordinal encoding to the age just by giving the numbers. 
#the above process is also called a target guiding. 
```

target guiding" refers to the process of using the target variable (also known as the dependent variable or response variable) to guide or inform the analysis, modeling, or decision-making process. The target variable is the variable that you are trying to predict or explain using other variables in a given context

```python
df.head(100)
```

```python
#handling categorical feature city_category
#using dummies
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```
